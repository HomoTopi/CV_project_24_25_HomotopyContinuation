```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# Random gaussian matrix 3x3
norms = []
iterations = 100000
sigma = 1
for i in range(iterations):
    Eps = np.random.normal(0, sigma, (3, 3))
    norms.append(np.linalg.norm(Eps))

print('Mean:', np.mean(norms))
print('Std:', np.std(norms))

print('Theoretical mean:', 3 * sigma)
print('Theoretical std:', np.sqrt(3 * 2**0.5 * sigma**2))

sns.histplot(norms, bins=50, kde=True)
plt.title('Histogram of the Frobenius norm of a 3x3 Gaussian matrix')
plt.show()
```
