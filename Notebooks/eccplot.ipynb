# Plot eccentricity for the report


```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
theta = np.linspace(0, 2 * np.pi, 100)
for e in np.arange(0, 1.01, 0.1):
    a=1
    b = np.sqrt(1 - e**2)
    x = a * np.cos(theta)
    y = b * np.sin(theta)
    plt.plot(x, y, color=plt.cm.cool(e), alpha=0.8, label=f"e={e:.1f}")

sm = plt.cm.ScalarMappable(cmap='cool', norm=plt.Normalize(vmin=0, vmax=1))
ax = plt.gca()
cbar = plt.colorbar(sm, ax=ax, label='Eccentricity')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.title('Ellipses with varying eccentricity')
plt.show()
```
