```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1 = np.random.rand(3, 1)
v2 = np.random.rand(3, 1)
v1, v2
```

```python
C_star = v1 @ v2.T + v2 @ v1.T
C_star
```

```python
U, S, Vt = np.linalg.svd(C_star)

U, S, Vt
```

```python
iter = 10000

V1s = []
V2s = []
Ss = []
for i in range(iter):
    # Generate random vectors
    v1 = np.random.rand(3, 1)*2- 1
    v2 = np.random.rand(3, 1)*2 - 1

    # Save vectors for plotting
    V1s.append(v1)
    V2s.append(v2)

    # Compute C_star
    C_star = v1 @ v2.T + v2 @ v1.T

    # Perform SVD
    U, S, Vt = np.linalg.svd(C_star)

    # Store singular values
    Ss.append(S)

# Convert to numpy array for easier manipulation
V1s = np.array(V1s)
V2s = np.array(V2s)
Ss = np.array(Ss)
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 12))

# Singular Value 1
plt.subplot(3, 1, 1)
plt.hist(Ss[:, 0], bins=30, alpha=0.7, color='blue')
plt.title('Distribution of Singular Value 1')
plt.xlabel('Singular Value')
plt.ylabel('Frequency')

# Singular Value 2
plt.subplot(3, 1, 2)
plt.hist(Ss[:, 1], bins=30, alpha=0.7, color='green')
plt.title('Distribution of Singular Value 2')
plt.xlabel('Singular Value')
plt.ylabel('Frequency')

# Singular Value 3
plt.subplot(3, 1, 3)
plt.hist(Ss[:, 2], bins=30, alpha=0.7, color='red')
plt.title('Distribution of Singular Value 3')
plt.xlabel('Singular Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
